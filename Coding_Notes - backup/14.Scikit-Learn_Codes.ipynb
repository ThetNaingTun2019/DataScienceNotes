StandardScalar() သည် fit လုပ်ပြီးမှ သိမ်းြခင်းပြုလုပ်နိုင်သည်။ StandardScalar သည် x-U/S ဖြစ်သည်။ mean ကိုဖယ်ပြီး std deviation ဖြင့်စားထားခြင်းဖြစ်သည်။

```
from sklearn import preprocessing
sc_X=preprocessing.StandardScaler() #ဒီအတိုင်းက function ခေါ်တာပဲရှိသေးတယ်။ pickle လုပ်လို့ ဘာမှသုံးလို့မရသေး။
X=[[1,2,3,4,5,6,7,8,9]]   #X သည် 2D Array ဖြစ်ရန်လို
X_std = sc_X.fit_transform(X)
sc_X.mean_  #StandardScalar ၏ mean ကိုရှာခြင်း
X_std.mean(axis=1)
```

```
Y=[[4,5,6,7,8,9,10,11,12]]
Y_std=sc_X.transform(Y)
Y_std
```
