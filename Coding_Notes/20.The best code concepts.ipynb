**How to count for each item? In Efficient way better than Michigan course**





```
fruits=['apple','banana','cherry','apple','banana','banana']
fruit_qty={}
for each in fruits:
  fruit_qty[each]=fruit_qty.get(each,0)+1

fruit_qty
```
