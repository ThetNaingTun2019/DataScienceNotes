```
images=[]
labels=[]
for each_channel in range(3):
  for each_intensity in [0.1,0.5,0.9]:
      split_channel=image.split() #to split each channel
      mid = split_channel[each_channel].point(lambda x:x*each_intensity) #to prepare intensity changes
      split_channel[each_channel].paste(mid) #to paste or apply changes intensity
      conv_img = Image.merge(image.mode, split_channel) #combines them into a single image
      images.append(conv_img)
      labels.append('channel {} intensity {}'.format(each_channel,each_intensity))
font = ImageFont.truetype("/content/fanwood-webfont.ttf",75)
single_image=images[0]
contact_sheet=PIL.Image.new(single_image.mode, (single_image.width*3,single_image.height*3+3*75))
x=0
y=0
draw = ImageDraw.Draw(contact_sheet)

for i,img in enumerate(images):
    contact_sheet.paste(img, (x, y))
    draw.text((x,y+single_image.height+15), labels[i], font=font)
    if x+single_image.width == contact_sheet.width:
        x=0
        y=y+single_image.height+75
    else:
        x=x+single_image.width  

contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))
display(contact_sheet)
```
