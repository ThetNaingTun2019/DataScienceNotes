##Bootstrap Distribution##

Bootstrap က ရှိတဲ့ sample ကို full 100% random ကို with replacement နဲ့ ယူတာပါ replicate အကြိမ်ယူသလောက် နဂို sample ထက်ကြီးလာမှာပါ

Data တွေ အကုန်မကောက်နိုင်တဲ့အခါ sample ကနေ population ကို ပြန်သွားရတာပါ

***Note—*** Bootstrapping cannot correct bias from sampling.

The mean of the bootstrap distribution will always be near the sample mean, which may not necessarily be very close to the population mean.

35% of software developers started programming as children.

A *hypothesis* is a statement about an unknown population parameter.

A *hypothesis test* is a test of two competing hypotheses.

*   The *null hypothesis (H0)* is the existing idea.
*   The *alternative hypothesis (HA)* is the new "challenger" idea of the researcher.



For our problem:

H0: The proportion of data scientists starting programming as children is 35%.

HA: The proportion of data scientists starting programming as children is greater than 35%.

Hypothesis tests check if the sample statistics lie in the tails of the null distribution.

p-values quantify evidence for the null hypothesis

Large p-value → fail to reject null hypothesis

Small p-value → reject null hypothesis

What decide small or large?
Now,** Statistical Significance** enter.

#Significance Level#

The significance level of a hypothesis test (α) is the threshold point for "beyond a
reasonable doubt"

Common valve of alpha are 0.2, 0.1, 0.05, and 0.01.

If p<=alpha, reject H0, else fail to reject H0.

alpha should be set prior to conducting the hypothesis test.
